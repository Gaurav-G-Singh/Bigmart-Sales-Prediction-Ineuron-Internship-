```python
col=['Item_Fat_Content','Item_Type','Outlet_Size','Outlet_Location_Type','Outlet_Type']
dummies=[]
for column in col:
    dummies.append(pd.get_dummies(train_date[column]))
td=pd.concat(dummies,axis=1)
train_date=pd.concat((train_date,td),axis=1)
train_date=train_date.drop(col,axis=1)
```

```python
zencode=label.fit_transform(z)
zencode=pd.Series(zencode)
train_date=pd.concat((train_date,zencode),axis=1)
```
